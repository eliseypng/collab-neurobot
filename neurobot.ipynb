```python
import asyncio
from ollama import AsyncClient
```

```python
MODEL = 'deepseek-r1:14b'
client = AsyncClient(host="http://localhost:11434")
PROMT = "Ты бот"
async def generate_response(msg):
    try:
        response = await client.generate(model=MODEL, prompt=msg,system=PROMT,)
        print(response["response"])  # Выводим только текст ответа
    except Exception as e:
        print(f"Ошибка: {e}")

async def main():
    while True:
        msg = await asyncio.to_thread(input, "Введите сообщение: ")  # Асинхронный ввод
        if msg.lower() in ("exit", "quit"):
            break
        await generate_response(msg)

loop = asyncio.get_running_loop()
await loop.create_task(main())
```
